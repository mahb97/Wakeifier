```python
def modify_embedding_with_seed(embedding, seed_pair_embeddings):
    if len(seed_pair_embeddings) == 2:
        return embedding + (seed_pair_embeddings[1] - seed_pair_embeddings[0])
    else:
        return embedding

# the stuff that gives Molly nightmares
seed_word_list = [word for pair in seed_pairs for word in pair]
seed_embeddings = model.encode(seed_word_list)

seed_embedding_dict = {}
for i in range(0, len(seed_embeddings), 2):
    seed_embedding_dict[seed_pairs[i//2]] = (seed_embeddings[i], seed_embeddings[i+1])

original_embedding_0 = embeddings[0]
seed_pair_to_use = seed_pairs[0]
seed_pair_embeddings_to_use = seed_embedding_dict[seed_pair_to_use]

modified_embedding_0 = modify_embedding_with_seed(original_embedding_0, seed_pair_embeddings_to_use)

print(f"Original embedding for the first idea:\n{original_embedding_0[:10]}...")
print(f"\nModified embedding for the first idea (using {seed_pair_to_use}):\n{modified_embedding_0[:10]}...")



# random beds
random_embedding = np.random.rand(embeddings.shape[1])
all_embeddings_with_random = np.vstack((embeddings, random_embedding))

print(f"\nOriginal number of embeddings: {embeddings.shape[0]}")
print(f"Number of embeddings after adding a random one: {all_embeddings_with_random.shape[0]}")

new_concept = "syntactic sugar"
new_concept_embedding = model.encode([new_concept])[0]
all_embeddings_with_new_concept = np.vstack((all_embeddings_with_random, new_concept_embedding))

print(f"Number of embeddings after adding a random one and a new concept: {all_embeddings_with_new_concept.shape[0]}")
```

```python
from sklearn.metrics.pairwise import cosine_similarity
all_embeddings = np.vstack((embeddings, modified_embedding_0.reshape(1, -1), random_embedding.reshape(1, -1), new_concept_embedding.reshape(1, -1)))
print(f"Total number of embeddings to analyze: {all_embeddings.shape[0]}")

similarity_matrix = cosine_similarity(all_embeddings, embeddings)
nearest_original_indices = np.argmax(similarity_matrix, axis=1)

print("\nNearest original ideas for each embedding in our expanded set:")

labels = [f"Original Idea {i}" for i in range(embeddings.shape[0])] + \
         [f"Modified Idea 0 (using {seed_pair_to_use})"] + \
         ["Random Embedding"] + \
         [f"New Concept Embedding ('{new_concept}')"]

for i, nearest_index in enumerate(nearest_original_indices):
    print(f"{labels[i]}: -> '{df['fun_idea'].iloc[nearest_index]}'")
```

```python
def combine_seed_effects(embedding, seed_pair_embeddings_list):
    modified_embedding = embedding.copy()
    for seed_pair_embeddings in seed_pair_embeddings_list:
        if len(seed_pair_embeddings) == 2:
            modified_embedding += (seed_pair_embeddings[1] - seed_pair_embeddings[0])
    return modified_embedding

original_embedding_index = 1
seed_pair_1_to_use = seed_pairs[1]
seed_pair_2_to_use = seed_pairs[2]

original_embedding = embeddings[original_embedding_index]
seed_pair_embeddings_1 = seed_embedding_dict[seed_pair_1_to_use]
seed_pair_embeddings_2 = seed_embedding_dict[seed_pair_2_to_use]

combined_modified_embedding = combine_seed_effects(original_embedding, [seed_pair_embeddings_1, seed_pair_embeddings_2])

print(f"Original embedding for Idea {original_embedding_index}:\n{original_embedding[:10]}...")
print(f"\nModified embedding with combined effects of {seed_pair_1_to_use} and {seed_pair_2_to_use}:\n{combined_modified_embedding[:10]}...")

combined_similarity = cosine_similarity(combined_modified_embedding.reshape(1, -1), embeddings)
nearest_original_index_combined = np.argmax(combined_similarity)

print(f"\nNearest original idea to the combined modified embedding: -> '{df['fun_idea'].iloc[nearest_original_index_combined]}'")
```

```python
def targeted_embedding_modification(embedding, target_embedding, direction='towards', strength=0.5):
    direction_vector = target_embedding - embedding

    if direction == 'away':
        direction_vector *= -1

    modified_embedding = embedding + direction_vector * strength
    return modified_embedding

original_embedding_index_target = 0
target_concept = "chaos"
target_concept_embedding = model.encode([target_concept])[0]

original_embedding_target = embeddings[original_embedding_index_target]

nudged_embedding_towards = targeted_embedding_modification(original_embedding_target, target_concept_embedding, direction='towards', strength=0.7) # Increased strength for more chaos!

print(f"Original embedding for Idea {original_embedding_index_target}:\n{original_embedding_target[:10]}...")
print(f"\nEmbedding nudged towards '{target_concept}':\n{nudged_embedding_towards[:10]}...")

away_concept = "boring"
away_concept_embedding = model.encode([away_concept])[0]

nudged_embedding_away = targeted_embedding_modification(original_embedding_target, away_concept_embedding, direction='away', strength=0.7)

print(f"\nEmbedding nudged away from '{away_concept}':\n{nudged_embedding_away[:10]}...")


nudged_embeddings = np.vstack((nudged_embedding_towards.reshape(1, -1), nudged_embedding_away.reshape(1, -1)))
nudged_similarity = cosine_similarity(nudged_embeddings, embeddings)
nearest_original_indices_nudged = np.argmax(nudged_similarity, axis=1)

print(f"\nNearest original idea to the embedding nudged towards '{target_concept}': -> '{df['fun_idea'].iloc[nearest_original_indices_nudged[0]]}'")
print(f"Nearest original idea to the embedding nudged away from '{away_concept}': -> '{df['fun_idea'].iloc[nearest_original_indices_nudged[1]]}'")
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

all_embeddings_for_viz = np.vstack((embeddings,
                                   modified_embedding_0.reshape(1, -1),
                                   random_embedding.reshape(1, -1),
                                   new_concept_embedding.reshape(1, -1),
                                   combined_modified_embedding.reshape(1, -1),
                                   nudged_embedding_towards.reshape(1, -1),
                                   nudged_embedding_away.reshape(1, -1)
                                  ))

pca = PCA(n_components=2)
reduced_embeddings = pca.fit_transform(all_embeddings_for_viz)

labels_for_viz = [f"Original {i}" for i in range(embeddings.shape[0])] + \
                 [f"Modified ('{seed_pair_to_use}')"] + \
                 ["Random"] + \
                 [f"New Concept ('{new_concept}')"] + \
                 [f"Combined Mods ('{seed_pair_1_to_use}', '{seed_pair_2_to_use}')"] + \
                 [f"Nudged Towards ('{target_concept}')"] + \
                 [f"Nudged Away ('{away_concept}')"]

plt.figure(figsize=(10, 8))
scatter = plt.scatter(reduced_embeddings[:, 0], reduced_embeddings[:, 1])

for i, label in enumerate(labels_for_viz):
     plt.annotate(label, (reduced_embeddings[i, 0], reduced_embeddings[i, 1]), textcoords="offset points", xytext=(0,10), ha='center')

plt.title("2D PCA Projection of Embeddings (Original, Modified, and Added)")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.grid(True)
plt.show()

# For better label placement with many points, you might need to install and use adjustText:
!pip install adjustText
from adjustText import adjust_text
texts = []
for i, label in enumerate(labels_for_viz):
     texts.append(plt.text(reduced_embeddings[i, 0], reduced_embeddings[i, 1], label))
adjust_text(texts)

print("\nInterpretation of the visualization:")
print("Each point represents an embedding. Points that are closer together in this 2D space are generally more similar in the original high-dimensional space.")
print("Observe how the modified and added embeddings are positioned relative to the original embeddings.")
print("Are the nudged embeddings closer to their target concepts? Do the combined modifications push embeddings in unexpected directions?")
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

all_embeddings_analysis = np.vstack((embeddings,
                                     modified_embedding_0.reshape(1, -1),
                                     random_embedding.reshape(1, -1),
                                     new_concept_embedding.reshape(1, -1),
                                     combined_modified_embedding.reshape(1, -1),
                                     nudged_embedding_towards.reshape(1, -1),
                                     nudged_embedding_away.reshape(1, -1)
                                    ))

labels_analysis = [f"Original {i}" for i in range(embeddings.shape[0])] + \
                  [f"Modified ('{seed_pair_to_use}')"] + \
                  ["Random"] + \
                  [f"New Concept ('{new_concept}')"] + \
                  [f"Combined Mods ('{seed_pair_1_to_use}', '{seed_pair_2_to_use}')"] + \
                  [f"Nudged Towards ('{target_concept}')"] + \
                  [f"Nudged Away ('{away_concept}')"]


# Cosine distance = 1 - Cosine Similarity

similarity_matrix_analysis = cosine_similarity(all_embeddings_analysis, embeddings)
cosine_distance_matrix = 1 - similarity_matrix_analysis

print("Cosine Distance to Nearest Original Idea:")
for i in range(embeddings.shape[0], len(labels_analysis)):
    nearest_original_index = np.argmax(similarity_matrix_analysis[i])
    distance_to_nearest = cosine_distance_matrix[i, nearest_original_index]
    print(f"{labels_analysis[i]}: Distance to Original {nearest_original_index} -> {distance_to_nearest:.4f}")

print("\nLook at that distance.")

N = 3

print(f"\nTop {N} Nearest Original Ideas for Modified/Added Embeddings:")

for i in range(embeddings.shape[0], len(labels_analysis)):
    top_n_indices = np.argsort(similarity_matrix_analysis[i])[::-1][:N]
    print(f"{labels_analysis[i]}:")
    for rank, original_index in enumerate(top_n_indices):
        similarity_score = similarity_matrix_analysis[i, original_index]
        print(f"  Rank {rank + 1}: Original {original_index} (Similarity: {similarity_score:.4f}) -> '{df['fun_idea'].iloc[original_index]}'")

print("\nSemantic fields are like puddles")
```

```python
def interpolate_embeddings(embedding_start, embedding_end, num_steps):
    interpolated_embeddings = []

    direction_vector = embedding_end - embedding_start

    for i in range(num_steps):
        step = i / (num_steps - 1) if num_steps > 1 else 0
        interpolated_embedding = embedding_start + direction_vector * step
        interpolated_embeddings.append(interpolated_embedding)

    return np.array(interpolated_embeddings)

embedding_start = embeddings[0]

embedding_end = modified_embedding_0

num_interpolation_steps = 10
interpolated_embeddings = interpolate_embeddings(embedding_start, embedding_end, num_interpolation_steps)

print(f"Generated {num_interpolation_steps} interpolated embeddings between Original Idea 0 and its modification:")
print(interpolated_embeddings.shape)

similarity_matrix_interp = cosine_similarity(interpolated_embeddings, embeddings)
nearest_original_indices_interp = np.argmax(similarity_matrix_interp, axis=1)

print("\nNearest original ideas along the interpolation path:")

# Display the nearest original idea for each interpolated point
for i, nearest_index in enumerate(nearest_original_indices_interp):
    print(f"Step {i + 1}/{num_interpolation_steps}: Nearest Original Idea {nearest_index} -> '{df['fun_idea'].iloc[nearest_index]}'")
```
